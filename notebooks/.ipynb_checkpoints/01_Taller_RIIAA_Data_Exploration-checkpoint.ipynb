```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
archivo = "../data/raw/nombres_DATA.txt"
nombres = np.genfromtxt(archivo,max_rows=1,dtype="U",delimiter=",")
# for i in enumerate(nombres):
#     print(i)
archivo = "../data/raw/DATA.csv"
data  = pd.read_csv(archivo,names=nombres,index_col=0,parse_dates=True,usecols=[0,20])
data
```

```python
# data.plot()
```

```python
archivo = '../data/raw/ESOLMET_2019.xlsx'
nombres=  ['tiempo','Ib',"Ig","Id","UV","To","HR","Ws","P"]
esolmet = pd.read_excel(archivo,skiprows=[0,1,2,3,4],index_col=0,parse_dates=True,names=nombres)
esolmet
```

```python
archivo = "../data/raw/RUOA.txt"
ruoa = pd.read_csv(archivo,skiprows=[0,2,3],index_col=0,parse_dates=True,sep="\t")
ruoa
```

```python
data = data.resample("600S").interpolate(method="time")
ruoa = ruoa.resample("600S").interpolate(method="time")
# esolmet
```

```python
data.plot()
ruoa.plot()
esolmet.plot()
```

```python
datos = pd.concat([data,ruoa,esolmet],axis=1)
# datos
```

```python
datos.columns
```

```python
fig, ax = plt.subplots(2,1,figsize=(10,8))

fecha1 = parse("2019-06-09")
fecha2 =  fecha1 + pd.Timedelta("10D")


ax[0].plot(datos.Ti,'r+')
ax[0].plot(datos.To,'k--')

ax[1].plot(datos.Ig,'g.')

ax[0].set_xlim(fecha1,fecha2)
ax[1].set_xlim(fecha1,fecha2)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from dateutil.parser import parse
```

```python
def DeltaMax(caso):
    return print(caso,(c.Ti.resample('D').max() - c[caso].resample('D').max()).mean())
```

```python
def DeltaMin(caso):
    return print(caso,(c.Ti.resample('D').min() - c[caso].resample('D').min()).mean())
```

```python
def DeltaFD(caso):
    FDe = (c.Ti.resample('D').max() - c.Ti.resample('D').min())/(c.To_esolmet.resample('D').max() - c.To_esolmet.resample('D').min())
    FDc  = (c[caso].resample('D').max() - c[caso].resample('D').min())/(c.To_esolmet.resample('D').max() - c.To_esolmet.resample('D').min())
    return print(caso,(FDe-FDc).mean())
    
```

```python

def importa(f,nombres):
    df = pd.read_csv(f,names=nombres,skiprows=1)
    df.tiempo = df.tiempo.str.replace("24:00:00","23:59:59")
    df.tiempo = "2019-" + df.tiempo
    df.tiempo = pd.to_datetime(df.tiempo,format="%Y- %m/%d %H:%M:%S")
    df.set_index('tiempo',inplace=True)
    return df
```

```python
f   = '../data/raw/bm.csv'
nombres = np.genfromtxt(f,max_rows=1,dtype='U',delimiter=',')
nombres[0] = 'tiempo'
nombres[9] = 'Ti'

bm = importa(f,nombres)
# bm
```

```python
dias = ['2019-06-09','2019-06-15']
caso = 'Ti_bm'
def DeltaLT(dias,caso):
    dt = pd.Timedelta(0)
    for dia in dias:
        dt = c.loc[dia].Ti.idxmax() - c.loc[dia][caso].idxmax() + dt
    dt = dt/len(dias)
    return dt
```

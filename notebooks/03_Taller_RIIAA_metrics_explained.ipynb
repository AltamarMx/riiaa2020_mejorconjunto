```python
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
```

```python
datos = pd.read_csv("../data/metrics/todos.csv",index_col=0,parse_dates=True)
datos
```

# Notaci'on
* $Ti$ Temperatura del aire al interior medida
* $Ti_{xx}$ Temperatura simulada del aire al interior de los caso ae, bm, ca, za
* $To_{esolmet}$ Temperatura del aire al exterior

```python
fig, ax = plt.subplots(figsize=(18,7))

fecha1 = parse("2019-06-22")
fecha2 = fecha1 + pd.Timedelta("2D")

ax.set_xlim(fecha1,fecha2)


ax.plot(datos.Ti_ae,'r+')
ax.plot(datos.Ti_bm,'g+')
ax.plot(datos.Ti_ca,'b+')
ax.plot(datos.Ti_za,'y+')

ax.plot(datos.Ti,'k-',linewidth=4)
```

# Consideraciones
* Datos medidos $M$
* Datos simulados $S$
* Datos cada 10 minutos
* Varios días de datos


# Métricas usadas

# Índice de Pearson $r$



#  $rmse$



# Diferencia promedio de la temperatura


$<\Delta T_i > = \frac{1}{N}  \sum_{i=1}^N (S_i-M_i)$

#  Diferencia promedio de las temperaturas máximas diarias
* $<\Delta Ti_{max} > =  \frac{1}{N} \sum_{i=0}^N (S_{max} - M_{max})$

#  Diferencia promedio de las temperaturas mínimas
* $<\Delta Ti_{min} > =  \frac{1}{N} \sum_{i=0}^N (S_{min} - M_{min})$


#  Diferencia promedio en el Factor de decremento
* $FD=\left( \frac{ T_{i_{max}}-T_{i_{min}}  }{T_{o_{max}}-T_{o_{min}}} \right )_S - \left( \frac{ T_{i_{max}}-T_{i_{min}}  }{T_{o_{max}}-T_{o_{min}}} \right )_M $

#   Diferencia promedio en el tiempo de retraso
* $<\Delta TR >  = \frac{1}{N} \sum_{d=0}^{N} (TR_{S_d} - TR_{M_d})$

donde

$TR_d = t(Ti_{max}) - t(To_{max})$

```python
fig, ax  = plt.subplots(figsize=(15,6),sharex=True)
fecha1 = parse("2019-06-22")
fecha2 = fecha1 + pd.Timedelta("2D")

ax.set_xlim(fecha1,fecha2)
ax.set_ylim(22,32)

# ax.plot(datos.To_esolmet,'r-',lw=3,label='To')
ax.plot(datos.Ti,        'k-',lw=3,label='Ti')


ax.plot(datos.Ti_ae,'ro:',label="Sim C1",markevery=10)
ax.plot(datos.Ti_bm,'bo:',label="Sim C2",markevery=10)
ax.plot(datos.Ti_ca,'ko:',label="Sim C3",markevery=10)
ax.plot(datos.Ti_za,'yo:',label="Sim C4",markevery=10)

ax.legend()
```
